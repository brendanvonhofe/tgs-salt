```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import os
```

```python
PATH = Path("/Users/brendan/Desktop/fastai/tgs_salt/data")
IMAGES = PATH/'images'
MASKS = PATH/'masks'
TEST = PATH/'test'
```

```python
fns = os.listdir(IMAGES)
test_fns = os.listdir(TEST)
fn = fns[1]
```

# Script to reflect pad images

```python
IMAGES128 = PATH/'images128'
MASKS128 = PATH/'masks128'
TEST128 = PATH/'test128'

ctr = 0
for filename in fns:
    ctr += 1
    if(ctr % 1000 == 0):
        print(ctr)
    abs_path = IMAGES/filename
    im = plt.imread(str(abs_path))
    im_128 = np.pad(im, ((0,27), (0,27), (0,0)), mode='reflect')
    plt.imsave(str(IMAGES128/filename), im_128)
```
